## **(e)** 

*Using the full data set, investigate the predictors graphically, using scatterplots or other tools of your choice. Create some plots highlighting the relationships among the predictors. Comment on your findings.*

```python
df = auto.rename(columns = {"mpg":"galonBasinaMil","cylinders":"silindir","displacement":"motorHacmi",
                       "horsepower":"beygirGücü","weight":"araçAğırlığı",
                       "acceleration":"ivme","year":"modelYılı",
                       "origin":"otomobilinMenşesi","name":"adı"})
```

```python
df.head()
```

```python
import seaborn as sns

sns.set(style="ticks")
sns.pairplot(df, kind="reg");
```

```python
sns.distplot(df["galonBasinaMil"],kde = True);
```

```python
sns.boxplot(x = "otomobilinMenşesi", y = "galonBasinaMil", data = df);
```

Otomobilin menşesinin Amerikan, Avrupa, Japon olmasına göre galon başına mil değişiyor.

```python
sns.boxplot(x = "silindir", y = "galonBasinaMil", data = df);
```

Silindir sayısı galon başına milde etkili bir tahmin edici olabilir.
